```python


import pandas as pd
import numpy as np

''' pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100) '''

df = pd.read_pickle("/content/drive/MyDrive/hcpa_ic.pkl")

display(df)

df['PESO'] = np.nan
df['PESO_DT'] = np.nan

df['ALTURA'] =  np.nan
df['ALTURA_DT'] = np.nan

df['PO2'] = np.nan
df['PO2_DT'] = np.nan

df['PCO2'] = np.nan
df['PCO2_DT'] = np.nan

df['QSOFA'] = np.nan
df['QSOFA_DT'] = np.nan

df['SpO2'] = np.nan
df['SpO2_DT'] = np.nan

df['FC'] = np.nan
df['FC_DT'] = np.nan

df['FR'] = np.nan
df['FR_DT'] = np.nan

df['PAS'] = np.nan
df['PAS_DT'] = np.nan

df['PAD'] = np.nan
df['PAD_DT'] = np.nan

i = 0
try:
  while True:
    colunas_valores = [df['PESO_contagem'][i], df['ALTURA_contagem'][i], df['PO2_contagem'][i], df['PCO2_contagem'][i], df['QSOFA_contagem'][i], df['SpO2_contagem'][i], df['FC_contagem'][i], df['FR_contagem'][i], df['PAS_contagem'][i], df['PAD_contagem'][i]]
    colunas_valores_int = []

    colunas_concat = [df['PESO_concat'][i], df['ALTURA_concat'][i], df['PO2_concat'][i], df['PCO2_concat'][i], df['QSOFA_concat'][i], df['SpO2_concat'][i], df['FC_concat'][i], df['FR_concat'][i], df['PAS_concat'][i], df['PAD_concat'][i]]

    colunas_sep = ['PESO', 'ALTURA', 'PO2', 'PCO2', 'QSOFA', 'SpO2', 'FC', 'FR', 'PAS', 'PAD']
    colunas_sep_dt = ['PESO_DT', 'ALTURA_DT', 'PO2_DT', 'PCO2_DT', 'QSOFA_DT', 'SpO2_DT', 'FC_DT', 'FR_DT', 'PAS_DT', 'PAD_DT']

    for j in range(len(colunas_valores)):
      if np.isnan(colunas_valores[j]) == False:
        colunas_valores_int.append(int(colunas_valores[j]))

    if max(colunas_valores_int,  default=0) > 1:
      for k in range(max(colunas_valores_int) - 1):
        valor_pos = ((i + k) + (i + k + 1)) / 2.0
        #print(valor_pos)
        df.loc[valor_pos] = np.nan
        df = df.sort_index().reset_index(drop=True)
        #df_vazio = pd.DataFrame({np.nan}, index=[i])
        #line = pd.DataFrame(index=[k])
        #df = pd.concat([df.iloc[:(k-1)], line, df.iloc[(k-1):]]).sort_index().reset_index(drop=True)

    if max(colunas_valores_int,  default=0) > 0:
      for l in range(len(colunas_concat)):
        valores_split = str(colunas_concat[l]).split(' | ')
        #print("valores_split", len(valores_split))
        for m in range(len(valores_split)):
          try:
            df.loc[(i + (m)), colunas_sep[l]] = valores_split[m].split('=')[1]
            df.loc[(i + (m)), colunas_sep_dt[l]] = valores_split[m].split('=')[0]
            ''' df.loc[(i + (m)), 'PRONTUARIO_2'] = df['PRONTUARIO_2'][(i - (m))]
            df.loc[(i + (m)), 'PRONTUARIO'] = df['PRONTUARIO'][(i - (m))] '''
          except:
            pass
    i += 1
except:
  pass

for i in range(len(df.index)):
  if np.isnan(df['PRONTUARIO'][i]) == True:
    df.loc[i, 'PRONTUARIO'] = df['PRONTUARIO'][i -1]

  if np.isnan(df['PRONTUARIO_2'][i]) == True:
    df.loc[i, 'PRONTUARIO_2'] = df['PRONTUARIO_2'][i -1]


display(df)
df.to_pickle('/content/drive/MyDrive/novo_hcpa_ic.pkl' )
```
